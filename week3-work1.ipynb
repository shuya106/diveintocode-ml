## 【課題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。

縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る  
全ての組み合わせの散布図を一度に表示させる、散布図行列を作る  
４つの特徴量に関する相関係数行列を作る  
相関係数行列のヒートマップを作る

```python
iris.target_names
```

```python
df.head(1)
```

```python
df_setosa = df[df["species"] == 0]
df_versicolor = df[df["species"] == 1]
df_virginica = df[df["species"] == 2]

plt.scatter(df_setosa["sepal length (cm)"], df_setosa["sepal width (cm)"], c="red", label="setosa")
plt.scatter(df_versicolor["sepal length (cm)"], df_versicolor["sepal width (cm)"], c="blue", label="versicolor")
plt.scatter(df_virginica["sepal length (cm)"], df_virginica["sepal width (cm)"], c="green", label="virginica")
plt.title("type of iris")
plt.xlabel("sepal length(cm)")
plt.ylabel("sepal width(cm)")
plt.legend()
```
